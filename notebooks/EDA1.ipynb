```python
import pandas as pd
import seaborn as sns
import numpy as np
df=pd.read_csv('data/finalTrain.csv')
```

```python
df=df.drop(labels=['Delivery_person_Age'],axis=1)
df=df.drop(labels=['Delivery_person_ID'],axis=1)
df=df.drop(labels=['ID'],axis=1)
import haversine as hs
hav=[]
for i in range(len(df)):
    loc1=(df.iloc[i,1],df.iloc[i,2])
    loc2=(df.iloc[i,3],df.iloc[i,4])
    hav.append(hs.haversine(loc1,loc2))
print(hav)
df['Distance']=hav
```

```python
df=df.drop(labels=['Restaurant_latitude'],axis=1)
df=df.drop(labels=['Restaurant_longitude'],axis=1)
df=df.drop(labels=['Delivery_location_longitude'],axis=1)
df=df.drop(labels=['Delivery_location_latitude'],axis=1)
df=df.drop(labels=['Order_Date'],axis=1)
df.info()
```

```python
res=df['Delivery_person_Ratings'].mean()
print(res)
df['Delivery_person_Ratings'].fillna(4.6)
df['Weather_conditions'].fillna('Fog')
df.mode()
```

```python
df['City']=df['City'].fillna('Metropolitian')
df['Festival']=df['Festival'].fillna('No')
df['multiple_deliveries']=df['multiple_deliveries'].fillna(1)
df['Road_traffic_density']=df['Road_traffic_density'].fillna('Low')
```

```python
import numpy as np
import datetime
converted_datetimes = []
for value in df['Time_Orderd']:
    if isinstance(value, str):  # Check if the value is a string
        try:
            converted_time = datetime.datetime.strptime(value, '%H:%M').time()
            converted_datetime = datetime.datetime.combine(datetime.datetime.min, converted_time)
        except ValueError:
            converted_datetime = None
    else:
        converted_datetime = None  # Handle NaN values
    converted_datetimes.append(converted_datetime)

df['time_column'] = converted_datetimes
# Extract numerical features from datetime.datetime column
df['Orderd_hour'] = df['time_column'].apply(lambda x: x.hour if x else None)
df['Orderd_minute'] = df['time_column'].apply(lambda x: x.minute if x else None)

# Drop the original datetime.datetime column
df.drop(columns=['time_column'], inplace=True)
```

```python
import numpy as np
converted_datetimes1 = []
for value in df['Time_Order_picked']:
    if isinstance(value, str):  # Check if the value is a string
        try:
            converted_time = datetime.datetime.strptime(value, '%H:%M').time()
            converted_datetime = datetime.datetime.combine(datetime.datetime.min, converted_time)
        except ValueError:
            converted_datetime = None
    else:
        converted_datetime = None  # Handle NaN values
    converted_datetimes1.append(converted_datetime)

df['Altered_Time_Order_picked'] = converted_datetimes1
# Extract numerical features from datetime.datetime column
df['Orderd_picked_hour'] = df['Altered_Time_Order_picked'].apply(lambda x: x.hour if x else None)
df['Orderd_picked_minute'] = df['Altered_Time_Order_picked'].apply(lambda x: x.minute if x else None)
df.drop(columns=['Altered_Time_Order_picked'], inplace=True)
```

```python
for index, row in df.iterrows():
    if pd.isna(row['Orderd_hour']) and pd.isna(row['Orderd_minute']):
        picked_time_in_minutes = row['Orderd_picked_hour'] * 60 + row['Orderd_picked_minute']
        adjusted_time_in_minutes = picked_time_in_minutes - 15
        
        adjusted_hour = adjusted_time_in_minutes // 60
        adjusted_minute = adjusted_time_in_minutes % 60
        
        df.at[index, 'Orderd_hour'] = adjusted_hour
        df.at[index, 'Orderd_minute'] = adjusted_minute
```
